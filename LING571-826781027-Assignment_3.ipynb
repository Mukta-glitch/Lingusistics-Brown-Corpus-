```python
from nltk.corpus import state_union
import pandas as pd
%matplotlib inline  
df = []
```

```python
for fileid in state_union.fileids():
    num_men = len([w for w in state_union.words(fileid) if w.upper()=='men'])
    num_women = len([w for w in state_union.words(fileid) if w.upper()=='women'])
    num_people = len([w for w in state_union.words(fileid) if w.upper()=='people'])
    year = int(fileid[:4])
    df.append([year, num_men, num_women, num_people])
print(df)
```

```python

df = pd.DataFrame(df, columns=('year', 'num_men', 'num_women', 'num_people'))
print(df.head())
df.plot(x='year', y=['num_men', 'num_women', 'num_people'],figsize=(12,8),grid=True)
```

```python
import nltk
cfd = nltk.ConditionalFreqDist((target,fileid[:4])
      for fileid in state_union.fileids() 
      for w in state_union.words(fileid) 
      for target in  ['men','women', 'people'] if w.lower().startswith(target))
cfd.plot()
```
